```python
from vits.infer import AudioInferencer
from IPython.display import Audio
```

```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
audioInferer = AudioInferencer("pretrained_models\Chameba.pth",device=device)
```

```python
text = ""
audio_data, sampling_rate = audioInferer.infer_audio(text=text,
                                                     speaker_id=42,
                                                     length_scale=1)
```

```python
import scipy.io.wavfile as wf
wf.write(filename="./tempFiles/temp1.wav",data=audio_data, rate=sampling_rate)
Audio(data = audio_data,rate = sampling_rate)
```
