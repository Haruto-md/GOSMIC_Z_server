```python
from vits.infer import AudioInferencer
from IPython.display import Audio
```

```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
audioInferer = AudioInferencer("pretrained_models\Rikyu.pth",device=device)
```

```python
text = "それぞれの国や地域で独自の文化として発展してきました"
audio_data, sampling_rate = audioInferer.infer_audio(text,43)
```

```python
import scipy.io.wavfile as wf
wf.write(filename="./tempFiles/temp1.wav",data=audio_data, rate=sampling_rate)
Audio(data = audio_data,rate = sampling_rate)
```
