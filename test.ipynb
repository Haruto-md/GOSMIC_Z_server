```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
from vits.infer import AudioInferencer
from IPython.display import Audio
device
```

```python
audioInferer = AudioInferencer("pretrained_models\Chameba.pth",device=device)
```

```python
text = "この音声は意図していない挙動を知らせるものです。キャラクター名が間違っています。"
audio_data, sampling_rate = audioInferer.infer_audio(text=text,
                                                     speaker_id=42,
                                                     length_scale=1)
```

```python
duration = 0.5
sliced_audio_data = audio_data[int(sampling_rate*duration):len(audio_data)-1-int(sampling_rate*duration)]
Audio(data = sliced_audio_data,rate = sampling_rate)
```

```python
import scipy.io.wavfile as wf
wf.write("tempFiles\\error_no_such_chara.wav",rate=sampling_rate,data=sliced_audio_data)
```
